Produce ice at 0°C from water at 20°C

Condenser temperature = T3 = T4 = 298K

Evaporator Temperature = T1 = T2 = 268K

COP_rel = 0.5

ṁ = 6 kg/min

x2 = 0.6

Cpw = 4.187 kj/kg K

hfg(ice) = 335 kj/kg

Time = t = 24 hrs

```python
T2 = T1 = 268
T3 = T4 = 298
sf2 = 0.1251
sf3 = 0.2232
hfg2 = 154.0
hfg3 = 138.0
hf2 = 31.4
hf3 = 59.7
x2 = 0.6
COP_rel = 0.5
m_dot = 6
Cpw = 4.187
hfs_ice = 335

""" We Know that process 2 ---> 3 will be isentropic
Therefore s2 = s3
"""

# sf2 + x2*hfg2/T2 = sf3 + x3*hfg3/T3
x3 = ((sf2 - sf3) + x2*hfg2/T2)*(T3/hfg3)
print("x3: ",x3)

h2 = hf2 + x2*hfg2
h3 = hf3 + x3*hfg3
print("h2: ",h2)
print("h3: ",h3)

# Considering that h4 = h1
h1 = h4 = 59.7

W = h3-h2
Rn = h2 - h1

COP_the = Rn/W
COP_act = COP_the*COP_rel

print('COP_act: ',COP_act)

# Now to find heat from 1kg of water at 20C to form ice at 0C
h_req = Cpw*(20-0) + hfs_ice

""" Now we know that COP_act = Rn/W and in our case
Rn(kj) = m_dot(ice) * h_req
W(kj) = m_dot(freon)* (h3 - h2)
"""
m_ice_dot = COP_act*W*m_dot/(h_req)

print('m(ice)_dot: ',m_ice_dot)

print('m(ice) in tonnes per day = ',m_ice_dot*60*24/1000)
```
