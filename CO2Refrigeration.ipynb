### Example 3.2


25 to -5°C

Dryness fraction of CO2 gas before suction stroke = 0.6

COP = 0.45

Temp_H2O_In = 15°C
Cp_H2O = 4.18 kJ/kg°C
LatHeat_H2O_solid = 335 kJ/kg

m_CO2_dot = 8.2 kg per minute (or) 0.1367 kg per second

Looking from perspective of Heat Pump/Refrigerator

Condenser = Heat is Rejected | Heat from Refrigerant is transfered to the Surroundings (Gas to Liq)
Evaporator = Heat is Accepted | Heat from Refrigerated Space to Vaporising Refrigerant (Liq to Gas)

**Objective:** *How many tonnes of ice at COP = 0.45 in 24 hours?*

```python
# Properties of CO2
sf2 = -0.042 #[kJ/kg K] #Entropy of Fluid
hfg2 = 245.2 #[kJ/kg] #Latent heat
hf2 = -7.53 #[kJ/kg] #Enthalpy of Fluid

sf3 = 0.251 #[kJ/kg K] #Entropy of Fluid
hfg3 = 121.34 #[kJ/kg] #Latent heat
hf3 = 81.17 #[kJ/kg] #Enthalpy of Fluid


T3 = 25 + 273.15 #[K]
T4 = T3

T1 = -5 + 273.15 #[K]
T2 = T1

x2 = 0.6

m_CO2_Refrig_dot = 0.1367 #[kg/s]

CoeffOfPerf_Relative = 0.45

# s2 = s3

# sf2 + (x2 * hfg2)/T2 = sf3 + (x3 * hfg3)/T3

# (x3 * hfg3)/T3 = sf2 + (x2 * hfg2)/T2 - sf3

x3 = ((sf2 + (x2 * hfg2)/T2 - sf3) * T3) / hfg3

h3 = hf3 + (x3 * hfg3)
print(h3)

h2 = hf2 + (x2 * hfg2)
print(h2)

W = h3 - h2
print(W)

# At h1, the refirgerant is fully in fluid condition and also enthalpy between h4 and h1 is constant. Therefore, h1 = hf4.

h1 = hf4 = hf3 # Although it doesnot make sense in conventional terms.

R_effect = h2 - h1

CoeffOfPerf_Theo = R_effect / W
print(CoeffOfPerf_Theo)

# CoeffOfPerf_Relative = CoeffOfPerf_Actual / CoeffOfPerf_Theo
CoeffOfPerf_Actual = CoeffOfPerf_Theo * CoeffOfPerf_Relative
print(CoeffOfPerf_Actual)

HeatExtractedPerSec = CoeffOfPerf_Actual * (m_CO2_Refrig_dot * (h3 - h2))
print(HeatExtractedPerSec)

HeatExtractedPerDay = HeatExtractedPerSec * 60 * 60 * 24 # [kJ/day]
print(HeatExtractedPerDay)

# Water is coming in at 15 degC. Phase change from Liquid to Solid

LatHeat_H2O = 335 #[kJ/kg]
C_p_H2O = 4.18 #[kJ/kg K]

HeattobeRemovedfromH2O = (C_p_H2O * (15 - 0)) + LatHeat_H2O # [kJ/kg K] * [K] = [kJ/kg]

print(HeattobeRemovedfromH2O)

Mass_of_H2O_Sol = HeatExtractedPerDay / HeattobeRemovedfromH2O # [kg/day]

print(Mass_of_H2O_Sol)
```
